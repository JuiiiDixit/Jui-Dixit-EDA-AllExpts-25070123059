```python
import pandas as pd
data = {
    "Name": ["A", "B", "C"],
    "Marks": [85, 90, 78]
}
df = pd.DataFrame(data)
df
```

```python
#Finding structure of data frame
print(df.shape)
print(df.ndim)
print(df.size) #3X2 rowsXcolumns as given in earlier code
```

```python
#Display column names
print(df.columns)
```

```python
#Check dat types of each column
(df.dtypes)
```

```python
#Access individual column
df["Name"]
```

```python
df["Marks"]
```

```python
#Access specific row
df.loc[1] #1 indicates the first row
```

```python
#Add a new column
'''If marks = 80-90, First class
If marks >= 90, Distinction
If marks < 80-90, Second class'''

df["Grade"] = ["First class", "Distinction", "Second class"]
df
```

```python
#Update data in data frame
#Change marks of student A to 88
df.loc[0, "Marks"] = 88
df
```

```python
#Delete a column
df.drop("Grade", axis = 1, inplace = True)
df
```

```python
#Perform basic statistical analysis
mean1 = df["Marks"].mean()
mean1
```

```python
#Find the minimum and maximum value of the given data set
mini = df["Name"].min()
print(mini)
maxi = df["Marks"].max()
print(maxi)
```

```python
#apply condition (filtering)
#Display students scoring more than 80
df[df["Marks"]>80]
```
