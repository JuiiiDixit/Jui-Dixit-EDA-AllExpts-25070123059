```python
#Creation of sets
thisset = {"Kiwi", "Mango", "Banana"}
print(thisset)
```

```python
#Duplicates not allowed
thisset = {"Kiwi", "Mango", "Banana", "Kiwi"}
print(thisset)
```

```python
#The values true and 1 are considered the same value in sets and are treated as duplicates
thisset = {"Kiwi", "Mango", "Banana", True, 1, 2}
print(thisset)
```

```python
#A set with strings, integers and boolean values
thisset = {"abc", 34, True, 30, "male", 5.5}
print(thisset)
```

```python
#Once a set is created you cannot replace its items, but you can add new items
thisset = {"Kiwi", "Mango", "Banana"}
thisset.add("Grapes")
print(thisset)
```

```python
#Set operations
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print("Union:", A | B)
print("Intersection:", A & B)
print("Difference:", A - B)
print("Symmetric difference:", A^B)
print(B-A)
```

```python
'''Problem statement 1: A college event has a list of students who registered multiple times by mistake. Write a program using set to store participant names and display only unique participants'''
participants = ["Jeon", "Jin", "Hob", "Jin", "Tim", "Jim", "Tim"]
print(participants)
remove_extra_participants = set(participants)
print(remove_extra_participants)
```

```python
'''Problem statement 2: Three students choose elective subjects. Find the common subjects chosen by all students using set operations.'''
std1_elective = {"Maths", "Science", "Social Science"}
std2_elective = {"Maths", "English", "Social Science"}
std3_elective = {"Maths", "Science", "English"}
print("Common subjects:", std1_elective|std2_elective|std3_elective)
```

```python
'''Problem statement 3: One set contains students in the cricket club, another in football club. Find students who are:
1. In both clubs
2. In only one club'''
Cricket = {"John", "Joon"}
Football = {"John", "Joon", "Maria", "Jen"}
print("Students in both clubs:", Cricket & Football)
print("Students in only ne club:", Cricket ^ Football)
```

```python
'''Problem statement 4: Absent students
Given a set of students and a set of present students, find the absent students'''
students = {"Naye", "Jeong", "Momo", "Sana", "Jihyo", "Mina", "Dahy", "Chae", "Tzu"}
present_students = {"Naye", "Momo", "Jihyo", "Tzu"}
print("Absent students:", students - present_students)
```

```python
'''problem statement 5: Remove invalid entries:
A set contains valid course codes. Remove the discontinued course code from the set'''
courses = {"EDP", "DCLD", "AMT", "IKS", "CT", "PES", "IES"}
courses.discard("IES")
courses.remove("IKS")
print(courses)
```
