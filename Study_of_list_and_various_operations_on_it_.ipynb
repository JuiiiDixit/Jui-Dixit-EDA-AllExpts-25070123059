```python
#Slicing
hislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(hislist[2:5])
```

```python
hislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(hislist[:4])
print(hislist[3:])
```

```python
hislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon"]
print(hislist[-3:-1])
print(hislist[-3:])
```

```python
#Change item
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant"
print(thislist)
```

```python
#Append items
thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)
```

```python
#Insert items
thislist = ["apple", "banana", "cherry"]
thislist.insert(1, "orange")
print(thislist)
```

```python
#Extend list
thislist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical)
print(thislist)
```

```python
numbers = [10, 20, 30, 40, 50]
#Slicing examples
print(numbers[1:4])
print(numbers[:3])
print(numbers[-3:])
print(numbers[:])
```

```python
numbers = [3, 7, 1, 9, 4]
print(max(numbers))
print(min(numbers))
print(sum(numbers))
print(len(numbers))
numbers.sort()
print(numbers)
numbers.sort(reverse=True)
print(numbers)
print(sum(numbers)/len(numbers)) #Average
```
