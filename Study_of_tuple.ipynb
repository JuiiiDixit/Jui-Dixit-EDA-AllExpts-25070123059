```python
#Create a tuple and access tuple items
thistuple = ("Kiwi", "Papaya", "Mango", "Avocado")
print(thistuple[0])
```

```python
#We cannot change tuple items
thistuple = ("Kiwi", "Papaya", "Mango", "Avocado")
thistuple[1] = "cherry"
print(thistuple)
```

```python
#Append items is also not possible
thistuple = ("Kiwi", "Papaya", "Mango", "Avocado")
thistuple.append("Grapes")
print(thistuple)
```

```python
#Add tuple to a tuple
thistuple = ("Kiwi", "Papaya", "Mango", "Avocado")
y = ("Grapes",)
thistuple += y
print(thistuple)
```

```python
#Slicing a tuple
thistuple = ("Kiwi", "Papaya", "Mango", "Avocado")
print(thistuple[1::2])
```

```python
'''1. A student's exam result is stored in a tuple containing three fixed fields:
Subject name, marks obtained and grade.
Write a python program to:
Store the exam result using tuple
Unpack the tuple elements using seperate variables.
Display the subject, marks and grade
Check whether the student has scored 75 or more and print
"Distinction" if the condition is satisfied'''

Result = ("maths", 82, "A")
subject, marks, grade = Result
print(subject, marks, grade)
if marks >= 75:
  print("Distinction")
```

```python
'''An organization records an employee's daily attendance for a week
using tuple, where:
"P" represents present, "A" represents absent
Write a python program to:
Store the attendance record in tuple
Count and display total number of present days
Count and display total number of absent days
Check whether the employee was absent at least once
during the recorded period and display the appropriate message'''

Attendance = ("P", "P", "P", "A", "P", "P", "P")
Attendance.count("P")
Attendance.count("A")
if Attendance.count("A")>=1:
  print("Employee was absent at least once")
```
